**1.   Data Awal**

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
pd.options.mode.chained_assignment = None  #menghilangkan peringatan

file_url = 'https://drive.google.com/file/d/1zjQilkcbW9AlhC_NuRI-tkR2vPIYKV7G/view?usp=share_link'

file_id = file_url.split('/')[-2]
dwn_url = "https://drive.google.com/uc?id=" + file_id
df = pd.read_csv(dwn_url, sep=';')

jumlah_baris = df.shape[0]
df.head(jumlah_baris)
```

**2.   Menghitung jarak pada nominal/categorical**

```python
category = df[['outlook']]
jumlah_kolom = category.shape[1]

for index, row in category.iterrows():
    if category.iloc[13]['outlook'] == category.iloc[index]['outlook']:
      category.loc[index,'jarak'] = (jumlah_kolom - 1) / jumlah_kolom
    else:
      category.loc[index,'jarak'] = (jumlah_kolom - 0) / jumlah_kolom

category.head(jumlah_baris)
```

**3.   Menghitung jarak pada biner**

```python
biner = df[['humidity', 'windy', 'play']]
jumlah_kolom = biner.shape[1]

for index, row in biner.iterrows():
  if biner.iloc[13]['humidity'] == biner.iloc[index]['humidity']:
      biner.loc[index,'bin_humidity'] = 1
  else:
      biner.loc[index,'bin_humidity'] = 0
  if biner.iloc[13]['windy'] == biner.iloc[index]['windy']:
      biner.loc[index,'bin_windy'] = 1
  else:
      biner.loc[index,'bin_windy'] = 0
  if biner.iloc[13]['play'] == biner.iloc[index]['play']:
      biner.loc[index,'bin_play'] = 1
  else:
      biner.loc[index,'bin_play'] = 0

biner.head(jumlah_baris)
```

```python
biner_new = biner[['bin_humidity', 'bin_windy', 'bin_play']]

biner_new['q'] = 0
biner_new['r'] = 0
biner_new['s'] = 0
biner_new['t'] = 0
biner_new['jarak'] = 0

for index, row in biner_new.iterrows():
  if biner_new.loc[13,'bin_humidity'] == 1 and biner_new.loc[index,'bin_humidity'] == 1:
    biner_new.loc[index, 'q'] += 1
  if biner_new.loc[13,'bin_humidity'] == 0 and biner_new.loc[index,'bin_humidity'] == 1:
    biner_new.loc[index, 'r'] += 1
  if biner_new.loc[13,'bin_humidity'] == 1 and biner_new.loc[index,'bin_humidity'] == 0:
    biner_new.loc[index, 's'] += 1
  if biner_new.loc[13,'bin_humidity'] == 0 and biner_new.loc[index,'bin_humidity'] == 0:
    biner_new.loc[index, 't'] += 1

  if biner_new.loc[13,'bin_windy'] == 1 and biner_new.loc[index,'bin_windy'] == 1:
    biner_new.loc[index, 'q'] += 1
  if biner_new.loc[13,'bin_windy'] == 0 and biner_new.loc[index,'bin_windy'] == 1:
    biner_new.loc[index, 'r'] += 1
  if biner_new.loc[13,'bin_windy'] == 1 and biner_new.loc[index,'bin_windy'] == 0:
    biner_new.loc[index, 's'] += 1
  if biner_new.loc[13,'bin_windy'] == 0 and biner_new.loc[index,'bin_windy'] == 0:
    biner_new.loc[index, 't'] += 1

  if biner_new.loc[13,'bin_play'] == 1 and biner_new.loc[index,'bin_play'] == 1:
    biner_new.loc[index, 'q'] += 1
  if biner_new.loc[13,'bin_play'] == 0 and biner_new.loc[index,'bin_play'] == 1:
    biner_new.loc[index, 'r'] += 1
  if biner_new.loc[13,'bin_play'] == 1 and biner_new.loc[index,'bin_play'] == 0:
    biner_new.loc[index, 's'] += 1
  if biner_new.loc[13,'bin_play'] == 0 and biner_new.loc[index,'bin_play'] == 0:
    biner_new.loc[index, 't'] += 1
  
  biner_new.loc[index, 'jarak'] = (biner_new.loc[index, 'r'] + biner_new.loc[index, 's']) / (biner_new.loc[index, 'q'] + biner_new.loc[index, 'r'] + biner_new.loc[index, 's'] + biner_new.loc[index, 't'])

biner_new.head(jumlah_baris)
```

**4.   Menggabungkan jarak categorical dan jarak biner**

```python
total_jarak = category[['jarak']] + biner_new[['jarak']]
total_jarak.head(jumlah_baris)
```

**5.   Mengurutkan data berdasarkan jarak yang terkecil**

```python
df_new = df.copy()
df_new['jarak'] = total_jarak[['jarak']]
df_new = df_new.drop(df_new.index[-1]) #menghapus index terakhir karena itu dirinya sendiri
df_new = df_new.sort_values('jarak') #sorting dari nilai terkecil

df_new.head(jumlah_baris)
```
