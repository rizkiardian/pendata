6. Membuat variabel x sebagai array NumPy untuk memilih semua baris (:), dan kolom 0 hingga 3 (indeks kolom 0, 1, 2, dan 3) dari DataFrame df_iris. Kolom-kolom ini mewakili fitur-fitur numerik yaitu sepal length, sepal width, petal length, dan petal width. Kemudian `.values` digunakan untuk mengubah DataFrame yang dipilih menjadi array NumPy.

```python
x = df_iris.iloc[:, [0,1,2,3]].values
print(x)
```

7. Implementasi metode elbow dalam analisis cluster pada dataset iris menggunakan algoritma KMeans pada Python. Metode elbow digunakan untuk menentukan jumlah klaster yang optimal dalam sebuah dataset. Elbow method memplot nilai within-cluster sum of squares (WCSS) yang dihasilkan oleh KMeans pada setiap jumlah klaster yang berbeda.

> Tujuan dari grafik ini adalah untuk menemukan lokasi "siku" pada kurva yang menunjukkan perubahan signifikan dalam penurunan WCSS. Jumlah klaster yang dipilih harus berada sebelum "siku" pada kurva tersebut. Angka ini akan digunakan sebagai jumlah klaster optimal dalam analisis cluster berikutnya.

```python
wcss = []

for i in range(1, 11):
  kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')  
plt.show()
```

8. Melakukan clustering pada dataset iris menggunakan algoritma KMeans pada Python.

> Penjelasan tiap parameter:

* n_clusters=3: Menentukan jumlah cluster yang diinginkan, dalam hal ini 3.
* init='k-means++': Menentukan metode inisialisasi centroid awal. 'k-means++' adalah salah satu metode inisialisasi yang umum digunakan pada K-Means.
* max_iter=300: Menentukan maksimum iterasi yang dilakukan dalam satu run K-Means, dalam hal ini sebanyak 300 iterasi.
* n_init=10: Menentukan berapa kali algoritma akan dijalankan dengan centroid awal yang berbeda. Dalam hal ini, algoritma akan dijalankan sebanyak 10 kali dengan centroid awal yang berbeda.
* random_state=0: Menentukan seed untuk menghasilkan angka random yang konsisten setiap kali kode dijalankan.

> Setelah model K-Means dilatih pada data x, klasifikasi cluster untuk setiap datapoint dihasilkan dan disimpan dalam variabel y_kmeans.

```python
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(x)
```

9. Memvisualisasikan hasil clustering dari model K-Means pada data x.

```python
plt.scatter(x[y_kmeans==0, 0], x[y_kmeans==0, 1], s=100, c='red', label='iris-setosa')
plt.scatter(x[y_kmeans==1, 0], x[y_kmeans==1, 1], s=100, c='blue', label='iris-versicolour')
plt.scatter(x[y_kmeans==2, 0], x[y_kmeans==2, 1], s=100, c='green', label='iris-virginica')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='yellow', label='Centroids')
plt.legend()
```

## Kesimpulan

Berdasarkan visualisasi diatas clustering dibagi menjadi 3 bagian, yaitu iris-setosa, iris-versicolour, dan iris-virginica. Terlihat dari pembagian kelompok tersebut cukup merata, namun terlihat pada kelompok iris-setosa dengan iris-virginica ada hubungan kedekatan.
