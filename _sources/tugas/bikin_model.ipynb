# Contoh Bikin Model

```python
import pickle
import numpy as np
```

```python
# Memuat/load model dari file
with open('mlp_clf_model.pkl', 'rb') as file:
    mlp_clf_model_loaded = pickle.load(file)


# Buat data baru
new_data = np.array([[5,	4,	1,	1.5]])

# Memuat/load data normalisasi
with open('scaler.pkl', 'rb') as file:
    scaler_loaded = pickle.load(file)

# Scaling data baru menggunakan skalar yang telah dilatih sebelumnya
new_data_scaled = scaler_loaded.transform(new_data)

# Menggunakan model untuk melakukan prediksi pada data baru
y_pred = mlp_clf_model_loaded.predict(new_data_scaled)
print(y_pred)
```
