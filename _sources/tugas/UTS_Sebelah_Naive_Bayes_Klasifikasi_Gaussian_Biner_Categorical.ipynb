Naive Bayes pada dasarnya adalah sebuah algoritma klasifikasi yang berdasarkan pada teorema Bayes dengan mengasumsikan bahwa setiap fitur (atau atribut) pada data yang diamati adalah independen satu sama lain. Oleh karena itu, algoritma ini dianggap "naif" karena tidak mempertimbangkan interaksi antara fitur.

Dalam klasifikasi, Naive Bayes menghitung probabilitas setiap kelas yang mungkin berdasarkan pada nilai fitur yang diamati, dan kemudian memilih kelas dengan probabilitas tertinggi sebagai prediksi. Algoritma ini sering digunakan dalam analisis teks, seperti pengklasifikasian email spam atau klasifikasi dokumen berdasarkan topik, tetapi juga dapat diterapkan pada berbagai jenis masalah klasifikasi lainnya.

1. import library pandas, numpy, dan sklearn yang diperlukan

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

2. import data kita

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasets/breast-cancer/master/data/breast-cancer.csv')

jumlah_baris = df.shape[0]
df.head(jumlah_baris)
```

3. Ubah nama fitur yang memiliki strip (-)

```python
df = df.rename(columns={'tumor-size': 'tumor_size'})
df = df.rename(columns={'inv-falsedes': 'inv_falsedes'})
df = df.rename(columns={'falsede-caps': 'falsede_caps'})
df = df.rename(columns={'deg-malig': 'deg_malig'})
df = df.rename(columns={'breast-quad': 'breast_quad'})
df = df.rename(columns={'class': 'class_name'})
df.head(jumlah_baris)
```

4. Menangani data hilang atau missing value

> dibawah ini menghitung data kosong:

```python
df_copy = df.copy()
df_copy.isnull().sum()
```

> ganti nilai yang mengandung tanda tanya (?) menjadi nilai kosong

```python
df_copy = df_copy.replace('?',np.nan)
```

```python
df_copy.isnull().sum()
```

5. Cek type data dari setiap fitur

```python
df_copy.info()
```

```python
df_copy['deg_malig'] = df_copy['deg_malig'].astype('object')
df_copy.info()
```

6. Ringkasan statistik dari semua atribut:

```python
df_copy.describe(include= 'all')
```

7. Langkah Imputasi Data atau mengisi data hilang dengan nilai yang didasarkan top dari ringkasan statistik diatas

```python
for value in ['falsede_caps', 'breast_quad']:
    df_copy[value].replace([None], [df_copy.describe(include='all')[value][2]], inplace=True)  #mengisi berdasarkan top describe

df_copy.head(jumlah_baris)
```

8. Pastikan sudah tidak ada data yang kosong lagi

```python
df_copy.isnull().sum()
```

9. One-Hot Encoder

> Encoder ini mengonversi data ke dalam format biner.

> Berikut adalah kode untuk menggunakan LabelEncoder pada beberapa kolom categorical pada dataset adult_df, dan kemudian melakukan encoding menggunakan One-Hot Encoder:

```python
le = preprocessing.LabelEncoder()
age = le.fit_transform(df_copy.age)
mefalsepause = le.fit_transform(df_copy.mefalsepause)
tumor_size = le.fit_transform(df_copy.tumor_size)
inv_falsedes = le.fit_transform(df_copy.inv_falsedes)
falsede_caps = le.fit_transform(df_copy.falsede_caps)
deg_malig = le.fit_transform(df_copy.deg_malig)
breast = le.fit_transform(df_copy.breast)
breast_quad = le.fit_transform(df_copy.breast_quad)
irradiat = le.fit_transform(df_copy.irradiat)
class_name = le.fit_transform(df_copy.class_name)
```

```python
#initialize the encoded categorical columns
df_copy['age'] = age
df_copy['mefalsepause'] = mefalsepause
df_copy['tumor_size'] = tumor_size
df_copy['inv_falsedes'] = inv_falsedes
df_copy['falsede_caps'] = falsede_caps
df_copy['deg_malig'] = deg_malig
df_copy['breast'] = breast
df_copy['breast_quad'] = breast_quad
df_copy['irradiat'] = irradiat
df_copy['class_name'] = class_name

df_copy.head(jumlah_baris)
```

10. Melakukan Standardisasi Data atau Standardization of Data

> Semua nilai data dari kerangka data adalah numerik. Sekarang, kita perlu mengonversinya dalam satu skala. Kita bisa menstandarkan nilai. Kita dapat menggunakan rumus di bawah ini untuk standardisasi:

```python
num_features = ['age', 'mefalsepause', 'tumor_size', 'inv_falsedes', 'falsede_caps', 'deg_malig', 'breast', 'breast_quad', 'irradiat']

scaled_features = {}
for each in num_features:
    mean, std = df_copy[each].mean(), df_copy[each].std()
    scaled_features[each] = [mean, std]
    df_copy.loc[:, each] = (df_copy[each] - mean)/std
```

11. Pemotongan Data atau Data Slicing

> Pisahkan data menjadi data training and data test. Kita dapat dengan mudah melakukan langkah ini menggunakan metode train_test_split() dari sklearn:

> Variabel features_train berisi data fitur untuk training, features_test berisi data fitur untuk test, target_train berisi target untuk training, dan target_test berisi target untuk test.

```python
features = df_copy.values[:,:9]  #Variabel features diisi dengan nilai features dari dataset df_copy dengan slicing [:,:9] artinya mengambil seluruh baris dan kolom 0 hingga kolom 3 (kolom ke-9 tidak diambil karena merupakan variabel target).
target = df_copy.values[:,9]  #Variabel target diisi dengan nilai target dari dataset df_copy dengan slicing [:,9] artinya mengambil seluruh baris dan hanya kolom ke-9 (variabel target).
features_train, features_test, target_train, target_test = train_test_split(features,target, test_size = 0.33, random_state = 10)  #Kemudian menggunakan fungsi train_test_split, dataset yang sudah dipisahkan dijadikan argumen, dengan tambahan argumen test_size = 0.33 yang berarti dataset akan dibagi menjadi 33% data uji dan 67% data latih, dan random_state = 10 sebagai seed untuk menghasilkan hasil yang sama setiap kali dijalankan.
```

12. Implementasi Gaussian Naive Bayes

> Setelah selesai preprocessing data. saatnya menerapkan algoritma pembelajaran mesin atau machine learning. Kita akan menggunakan modul GaussianNB sklearn:

```python
clf = GaussianNB()
clf.fit(features_train, target_train)
target_pred = clf.predict(features_test)
```

13. Akurasi model Gaussian Naive Bayes

```python
accuracy = accuracy_score(target_test, target_pred, normalize = True)

print(f"Akurasi model KNN: {accuracy:.2%}")
```
